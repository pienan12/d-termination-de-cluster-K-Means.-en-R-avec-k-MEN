```python
%%R

# Charger les bibliothèques R nécessaires
install.packages("factoextra", repos="http://cran.rstudio.com/")
library(factoextra)

# Lire les données
wine <- read.csv("wine.csv")

# Convertir la variable Type en facteur
wine$Type <- as.factor(wine$Type)

# Supprimer la première colonne (index) et la colonne Type pour la standardisation
wine_data <- wine[, -c(1, 2)]

# Standardiser les données
wine_scaled <- scale(wine_data)

# Déterminer le nombre optimal de clusters avec la méthode du coude
set.seed(123) # Pour la reproductibilité
wss <- sapply(1:10, function(k) {
  kmeans(wine_scaled, centers = k, nstart = 25)$tot.withinss
})

# Tracer la méthode du coude
plot(1:10, wss, type = "b", pch = 19, frame = FALSE,
     xlab = "Nombre de clusters (k)", ylab = "Somme des carrés intra-clusters (WSS)",
     main = "Méthode du coude pour déterminer le nombre optimal de clusters")

# Appliquer K-Means avec le nombre optimal de clusters
optimal_k <- 3 # À partir du graphique, on choisit 3 clusters
kmeans_result <- kmeans(wine_scaled, centers = optimal_k, nstart = 25)

# Afficher la taille des clusters
cluster_sizes <- table(kmeans_result$cluster)
print(cluster_sizes)

# Visualiser les clusters
fviz_cluster(kmeans_result, data = wine_scaled,
             geom = "point",
             ellipse.type = "convex",
             ggtheme = theme_minimal())
```
